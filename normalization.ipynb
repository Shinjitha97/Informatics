<h2>Code Count for Drug Mentions in UMLS with Chemical Disease Relation Codes  </h2>

```python
drug_UMLS = pd.read_pickle("./sem_eval_drug.pkl")
drug_UMLS = drug_UMLS.drop(columns=['DrugSynonyms_CDR'])
drug_UMLS = drug_UMLS.reset_index()
drug_UMLS = drug_UMLS.groupby(by='DrugsynonymsUMLS').agg('count')
drug_UMLS = drug_UMLS.rename(columns={"DrugsynonymsUMLS" : "Drug_Mentions(UMLS)","Code": "Code_count"})
drug_UMLS = drug_UMLS.reset_index()
drug_UMLS = drug_UMLS.to_excel("cdrdisease.xlsx") 
```
